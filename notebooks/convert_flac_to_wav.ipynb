```python
flac_files = glob.glob("../data/processed/flac/*.flac")
flac_files
```

```python
for i in flac_files:
    print(i)
    file_name = Path(i).name.replace("flac", "wav")
    song = AudioSegment.from_file(i, format="flac")
    song.export(f"../data/raw/negatives/{file_name}",format = "wav")
```
